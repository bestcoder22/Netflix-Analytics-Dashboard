# Split the 'Genre' column into multiple rows

```python
df.assign(Genre = df['listed_in'].str.split(', ')).explode('Genre')
```

```python
new_df = df.assign(Genre = df['listed_in'].str.split(', ')).explode('Genre')

new_df.head()
```

## Reset the Index

```python
new_df = new_df.reset_index(drop=True)

new_df.head()
```

## Group by 'Genre' and count occurrences
### Top Genre

```python
genre_count = new_df['Genre'].value_counts().reset_index()
genre_count.columns = ['Genre', 'Count']

genre_count.head()
```

## by 'Type'
### Top Genre by Type

```python
# Separate Movies and TV Shows into two DataFrames
movies_df  = new_df[new_df['type'] == 'Movie']
tv_shows_df = new_df[new_df['type'] == 'TV Show']
```

```python
movie_genre_count = movies_df['Genre'].value_counts().reset_index()
movie_genre_count.columns = ['Genre', 'Count']

movie_genre_count.head()
```

```python
tv_show_genre_count = tv_shows_df['Genre'].value_counts().reset_index()
tv_show_genre_count.columns = ['Genre', 'Count']

tv_show_genre_count.head()
```

# Split the 'cast' column into multiple rows

```python
new_df.assign(cast = new_df['cast'].str.split(', ')).explode('cast')
```

```python
new_df = new_df.assign(cast = new_df['cast'].str.split(', ')).explode('cast')

new_df.head()
```

## Group by 'Cast' and count occurrences

```python
new_df['cast'].value_counts()
```

```python
cast_counts_per_title = new_df.groupby('cast')['title'].nunique().reset_index(name='Count')
cast_counts_per_title.columns = ['Cast', 'Count']
cast_counts_per_title = cast_counts_per_title.sort_values('Count', ascending=False)

cast_counts_per_title.head()
```

# Group by 'Genre' and count occurrences after explosion of cast
### Top Genre

```python
# Group by Title and Genre, then count unique genres per Title
genre_counts_per_title = new_df.groupby('Genre')['title'].nunique().reset_index(name='Count')
genre_counts_per_title.columns = ['Genre', 'Count']
genre_counts_per_title = genre_counts_per_title.sort_values('Count', ascending=False)

genre_counts_per_title.head()
```

```python
# Group by Title and Genre, then count unique genres per Title
genre_counts_per_title = tv_shows_df.groupby('Genre')['title'].nunique().reset_index()
genre_counts_per_title.columns = ['Genre', 'Count']
genre_counts_per_title = genre_counts_per_title.sort_values('Count', ascending=False)

genre_counts_per_title.head()
```
